```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

# Generate Dataset

X is a 2d array of points and y is a 1d array that indicates which cluster each point belongs to. We can plot the points as follows:

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=2, random_state=42)
```

```python
plt.scatter(X[:,0], X[:,1])
```

# Choosing Number of Clusters - Elbow Method


```python
inertia = []
for k in range(1,10):
    #Initialize KMeans with k clusters
    KMeansModel = KMeans(n_clusters=k)
    #fit the data
    KMeansModel.fit(X)
    inertia.append(KMeansModel.inertia_)
```

```python
plt.plot(range(1,10),inertia)
```

```python
KMeansModel = KMeans(n_clusters=3)
KMeansModel.fit(X)
prediction = KMeansModel.predict(X)
```
